## №1

```python
# толком не понял задание, сделал как понял

ratings = (
    pd.read_csv('ml-latest-small/ratings.csv')
        .groupby(['movieId'])
        .agg({'rating': 'mean'})
        .reset_index()
)

def get_class(rating):
    try:
        if rating <= 2:
            return 'low'
        elif rating <= 4:
            return 'mid'
        else:
            return 'high'
    except:
        return 'n/a'

movies = (
    pd.read_csv('ml-latest-small/movies.csv')
        .merge(ratings, on='movieId', how='left')
)

movies['class'] = movies['rating'].apply(get_class)

movies.sort_values('rating', ascending=False).head()
```

# №2

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

def get_region(keyword):
    for region, cities in geo_data.items():
        for city in cities:
            if city in keyword:
                return region
            else:
                return 'undefined'

keywords = pd.read_csv('ml-latest-small/keywords.csv')

keywords['region'] = keywords['keyword'].apply(get_region)

keywords.head()
```

# №3

```python
def get_production_year(title):
    for year in range(1950, 2020):
        if str(year) in title:
            return year

    return 1990

movies['year'] = movies['title'].apply(get_production_year)

movies.groupby(['year']).agg({'rating': 'mean'}).sort_values('rating', ascending=False).head()
```
