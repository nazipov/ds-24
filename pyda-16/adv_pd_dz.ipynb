## Задание 1

```python
log = pd.read_csv('visit_log.csv', delimiter=';')

def get_source_type(row):
    traffic_source = row['traffic_source']
    region = row['region']

    if traffic_source == 'yandex' or traffic_source == 'google':
        return 'organic'
    elif traffic_source == 'paid' or traffic_source == 'email':
        if region == 'Russia':
            return 'ad'
        else:
            return 'other';
    else:
        return traffic_source

log['source_type'] = log.apply(get_source_type, axis=1)

log.head()
```

## Задание 3

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')

# не смог разобраться как одним groupby решить задау
user_life_time = (ratings
        .groupby('userId')
        .filter(lambda x: x['userId'].count() >= 100)
        .groupby('userId')['timestamp']
        .agg(lambda r: r.max() - r.min()).mean())

print(f"User life time = {user_life_time}")
```
