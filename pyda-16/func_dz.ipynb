```python
# !!! как бы нужно проверять уникальность номер+тип и так далее, но ничего не сказано в задании, поэтому ингорируем

#
# хранилище
#
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

#
# вспомогательные функции
#
def find_document(number):
    filtered = list(filter(lambda i: i['number'] == number, documents))
    if filtered == []:
        return None
    else:
        return filtered[0]

def find_directory(document_number):
    return next((k for k, v in directories.items() if document_number in v), None)

def remove_document_from_directories(number):
    directory = find_directory(number)
    directories[directory].remove(number)

#
# функции пользовательского ввода
#
def input_document_number():
    return input('Введите номер документа:')

def input_directory_number():
    return input('Введите номер полки:')

def input_document_and_directory():
    document = {
        'number': input('Введите номер документа:'),
        'type': input('Введите тип документа:'),
        'name': input('Введите владельца документа:')
    }

    directory = input('Введите полку для хранения:')

    return document, directory

#
# функции выводы данных
#
def print_directories_list():
    return print(F'Текущий перечень полок: {", ".join(directories.keys())}')

def print_documents(title = True):
    if title:
        print('Текущий список документов:')

    for document in documents:
        directory = find_directory(document['number'])
        print(F"№: {document['number']}, тип: {document['type']}, владелец: {document['name']}, полка хранения: {directory}")
    
#
# обработчики команд
#

# узнать владельца документа по его номеру
def command_p():
    document = find_document(input_document_number())
    if document is not None:
        print(F"Владелец документа: {document['name']}")
    else:
        print("Документ не найден в базе")

# по номеру документа узнать на какой полке он хранится
def command_s():
    directory = find_directory(input_document_number())
    if directory is not None:
        print(F"Документ хранится на полке: {directory}")
    else:
        print("Документ не найден в базе")

# увидеть полную информацию по всем документам
def command_l():
    print_documents(title = False)

# добавить новую полку
def command_ads():
    directory_number = input_directory_number()

    if directory_number not in directories:
        directories[directory_number] = []
        print(F'Полка добавлена.', end = ' ')
    else:
        print(F'Такая полка уже существует.', end = ' ')
        
    print_directories_list()

# удалить существующую полку из данных (только если она пустая)¶
def command_ds():
    directory_number = input_directory_number()

    if directories[directory_number] == []:
        directories.pop(directory_number)
        print(F'Полка удалена.', end = ' ')
    else:
        print(F'На полке есть документа, удалите их перед удалением полки.', end = ' ')
        
    print_directories_list()

# добавить новый документ в данные
def command_ad():
    document, directory = input_document_and_directory()
    
    if directory in directories:
        documents.append(document)
        directories[directory].append(document['number'])
        print('Документ добавлен.', end = ' ')
    else:
        print('Такой полки не существует. Добавьте полку командой ads.')

    print_documents()

# может удалить документ из данных
def command_d():
    document = find_document(input_document_number())
    if document is not None:
        remove_document_from_directories(document['number'])
        documents.remove(document)

        print("Документ удален.")
    else:
        print("Документ не найден в базе")

    print_documents()

# переместить документ с полки на полку
def command_m():
    document = find_document(input_document_number())
    target_directory = input_directory_number()
    
    if target_directory not in directories:
        print('Такой полки не существует.', end = ' ')
        print_directories_list()
        return
    elif document is None:
        print('Документ не найден в базе.')
    else:
        remove_document_from_directories(document['number'])
        directories[target_directory].append(document['number'])

        print('Документ перемещен.')

    print_documents()

# список команд
commands = ['p', 's', 'l', 'ads', 'ds', 'ad', 'd', 'm']

while True:
    command = input('Введите команду:')

    if command in commands:
        globals()['command_' + command]()
        
        print('---')
    elif command == 'q':
        break;
```
