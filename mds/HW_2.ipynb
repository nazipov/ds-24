```python
A = [
  [-100, -100, 1],
  [-100,  100, 1],
  [ 100,  100, 1],
  [   0,    0, 1],
  [ 100, -100, 1],
  [-100, -100, 1]
]

def draw_matrix(matrix):
    x = matrix[:,0]
    y = matrix[:,1]

    plt.plot(x, y)
    plt.ylim([-200, 200])
    plt.xlim([-200, 200])

A = np.array(A)

draw_matrix(A)
```

```python
# Уменьшить объект в два раза

resize = np.array([[0.5,   0, 0],
                   [  0, 0.5, 0],
                   [  0,   0, 1]])

draw_matrix(A @ resize)
```

```python
# Повернуть на 130 градусов
angle = np.deg2rad(130)

rotate = np.array([[math.cos(angle), -math.sin(angle), 0],
                   [math.sin(angle),  math.cos(angle), 0],
                   [              0,                0, 1]])

draw_matrix(A @ rotate)
```

```python
# Отразить объект относительно прямой y=x

mirror = np.array([[ 0, 1, 0],
                   [-1, 0, 0],
                   [ 0, 0, 1]])

draw_matrix(A @ mirror)
```
